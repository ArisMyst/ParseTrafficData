```python
import pandas as pd
df = pd.read_csv('firstnormaldata20191026_DecisionTree.csv')
df.head()
```

```python
inputs = df.drop('CONGESTION',axis='columns')
```

```python
target = df['CONGESTION']
```

```python
from sklearn.preprocessing import LabelEncoder
le_TIME = LabelEncoder()
le_DAY = LabelEncoder()
le_STORES = LabelEncoder()
le_OSM_ID = LabelEncoder()
le_LARGER_THAN_200M = LabelEncoder()
le_KMH = LabelEncoder()
le_CATEGORY = LabelEncoder()
```

```python
inputs['TIME_n'] = le_TIME.fit_transform(inputs['TIME'])
inputs['DAY_n'] = le_DAY.fit_transform(inputs['DAY'])
inputs['STORES_n'] = le_STORES.fit_transform(inputs['STORES'])
inputs['OSM_ID_n'] = le_OSM_ID.fit_transform(inputs['OSM_ID'])
inputs['LARGER_THAN_200M_n'] = le_LARGER_THAN_200M.fit_transform(inputs['LARGER_THAN_200M'])
inputs['KMH_n'] = le_KMH.fit_transform(inputs['KMH'])
inputs['CATEGORY_n'] = le_CATEGORY.fit_transform(inputs['CATEGORY'])
```

```python
inputs
```

```python
inputs_n = inputs.drop(['TIME','DAY','STORES','OSM_ID','LARGER_THAN_200M','KMH','CATEGORY'],axis='columns')
```

```python
inputs_n
```

```python
target
```

```python
#feature_cols = ['TIME_n', 'DAY_n', 'STORES_n', 'OSM_ID_n','LARGER_THAN_200M_n','KMH_n','CATEGORY_n'] 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs_n, target, test_size=0.2, random_state=1)
```

```python
X_test
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier(class_weight=None, criterion='entropy', max_depth=None,
            max_features=None, max_leaf_nodes=None,
            min_impurity_decrease=0.0, min_impurity_split=None,
            min_samples_leaf=1, min_samples_split=2,
            min_weight_fraction_leaf=0.0, presort=False, random_state=None,
            splitter='best')
model.fit(X_train, y_train)
```

```python
model.score(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
